## VISUALIZING OUTLIERS

```python
import matplotlib.pyplot as plt
import seaborn as sns

#visualize outliers with a boxplot
sns.boxplot(df['ProductPrice'])
plt.show()

```

```python
# identify outliers based on the IQR 
Q1 = df['ProductPrice'].quantile(0.25)
Q3 = df['ProductPrice'].quantile(0.7)
IQR = Q3 - Q1
```

```python
# identify outliers based on the IQR
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
# identify rows with outliers
outliers = df[(df['ProductPrice'] < lower_bound) | (df['ProductPrice'] > upper_bound)]
#print(outliers)
```

```python
# remove outliers
df_cleaned = df [(df['ProductPrice'] >= lower_bound) & (df['ProductPrice'] <=upper_bound)]
```

```python
# Cap outliers
df['ProductPrice'] = df['ProductPrice'].clip(lower=lower_bound, upper=upper_bound)
```
