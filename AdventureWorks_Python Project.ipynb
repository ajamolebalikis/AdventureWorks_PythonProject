## Perform Customer Segmentation Based on Purchasing Power and Spending Power


```python
import os
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# set the environment variable for MKL
os.environ["OMP_NUM_THREADS"] = "1"
```

```python
# merge sales_data and dimcustomer on customerkey
customer_sales = pd.merge(sales_data, Dimcustomer, on='CustomerKey', how='inner')
```

```python
# Perform customer segementation
segmentation_data = customer_sales[['TotalSales', 'Age', 'AnnualIncome']]
```

```python
# Drop any rows with missing data
segmentation_data = segmentation_data.dropna()
```

```python
# performs KMeans clustering to segment customers
import pandas as pd
from sklearn.cluster import KMeans

# Clean the column by removing '$' and ',' and convert to numeric
segmentation_data['AnnualIncome'] = (
    segmentation_data['AnnualIncome']
    .replace({r'\$': '', r',': ''}, regex=True)
    .astype(float)
)

# initialize the KMeans model(adjust the number of clusters as needed)
kmeans = KMeans(n_clusters=4)

# Ensure you pass only the numeric data to fit_predict
segmentation_data['clusters'] = kmeans.fit_predict(segmentation_data.select_dtypes(include='number'))


#Analyze clusters
print(segmentation_data.groupby('clusters').mean())

```

```python
# count the number of customers in each segment (cluster)
cluster_counts = segmentation_data['clusters'].value_counts()

# visualize the clusters based on TotalSales 
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.bar(cluster_counts.index, cluster_counts.values, color='skyblue')
plt.title('Number of Customers per Segment')
plt.xlabel('Customer Segment (clusters)')
plt.ylabel('Number of Customers')
plt.xticks(cluster_counts.index)
plt.show()
```

```python
# calculate the average total sales per cluster
average_sales_per_cluster = segmentation_data.groupby('clusters')['TotalSales'].mean()

#Visualization
plt.figure(figsize=(8, 6))
plt.bar(average_sales_per_cluster.index, average_sales_per_cluster.values, color='orange')
plt.title('Average Total Sales Per Customer Segment')
plt.xlabel('Customer Segment(clusters)')
plt.ylabel('Average Total Sales')
plt.xticks(average_sales_per_cluster.index)
plt.show()
```

## Perform Hypothesis Testing 

```python
print(sales_data['Region'].unique())
```

```python
# Compare the total sales for all regions
northwest_sales = sales_data[sales_data['Region'] =='Northwest']['TotalSales']
southwest_sales = sales_data[sales_data['Region'] =='Southwest']['TotalSales']
australia_sales = sales_data[sales_data['Region'] =='Australia']['TotalSales']
canada_sales = sales_data[sales_data['Region'] =='Canada']['TotalSales']
united_kingdom_sales = sales_data[sales_data['Region'] =='United_Kingdom']['TotalSales']
france_sales = sales_data[sales_data['Region'] =='France']['TotalSales']
germany_sales = sales_data[sales_data['Region'] =='Germany']['TotalSales']
southeast_sales = sales_data[sales_data['Region'] =='Southeast']['TotalSales']
northeast_sales = sales_data[sales_data['Region'] =='Northeast']['TotalSales']
central_sales = sales_data[sales_data['Region'] =='Central']['TotalSales']
```

```python
# List of all sales regions
Regions = {
    "Northwest": northwest_sales,
    "Southwest": southwest_sales,
    "Australia": australia_sales,
    "Canada": canada_sales,
    "United Kingdom": united_kingdom_sales,
    "France": france_sales,
    "Germany": germany_sales,
    "Southeast": southeast_sales,
    "Northeast": northeast_sales,
    "Central": central_sales,
}

# Check and print the data for all regions
for Region, Totalsales in Regions.items():
    print(f"{Region} Sales: {Totalsales}")
```

```python
# Perform a one-way ANOVA test
import scipy.stats as stats

f_stat, p_value = stats.f_oneway(northwest_sales, southwest_sales, australia_sales, canada_sales, france_sales, germany_sales, southeast_sales, northeast_sales, central_sales)
```

```python
# print the results
print(f"F-statistics: {f_stat}")
print(f"P-value: {p_value}")
```

```python
# Check if there is a significant difference between the regions
if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant difference in total sales amomg Northwest, Southwest, Australia, Canada, France, Germany, Southeast, Northeast, Central.")
else:
    print("Fail to reject the null hypothesis: There is no significant diference in total sales among Northwest, Southwest, Australia, Canada, France, Germany, Southeast, Northeast, Central.")
```

## Conduct Advanced Statistical Analysis


```python
# Perform linear regression to understand how Product Price affects Total Sales
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
# Get the data for regression analysis
X = sales_data[['ProductPrice']].values.reshape(-1, 1) # independent variables
Y = sales_data[['TotalSales']].values # dependent variables
```

```python
# Create and fit the regression model
model = LinearRegression()
model.fit(X, Y)
```

```python
# Get the regression coefficient and intercept
coef = model.coef_[0]
intercept = model.intercept_
```

```python
# Print the results
print(f"Regression coefficient: {coef}")
print(f"Intercept: {intercept}")
```

```python
# Predict total sales based on product price
Y_pred = model.predict(X)
```

```python
# Visualization

import matplotlib.pyplot as plt

# Plot the regression line
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue', label='Actual Sales')
plt.plot(X, Y_pred, color='red', label='Regression Line')
plt.title('Linear Regression: TotalSales vs. ProductPrice')
plt.xlabel('Product Price')
plt.ylabel('Total Sales')
plt.legend()
plt.show()
```
